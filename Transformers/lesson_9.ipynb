```python
def preprocces(text):
    text = text.lower()
    text = text.translate(str.maketrans("", "", string.punctuation))
    return text
```

```python
data = positive_sentence + negative_sentence
labels = [1] * len(positive_sentence) + [0] * len(negative_sentence)

data = [preprocces(text=sentence) for sentence in data]

all_word = " ".join(data).split()
word_count = Counter(all_word)
vocab = {word: idx+1 for idx, word in enumerate(word_count.items())}
vocab["<pad>"] = 0
```

```python
max_len = 42
def sentence_to_token(sentence, vocab, max_len):
    tokens = sentence.lower().split()
    indices = [vocab.get(token, vocab.get('<unk>', 0)) for token in tokens]
    if len(indices) < max_len:
        indices += [vocab.get('<pad>', 0)] * (max_len - len(indices))
    else:
        indices = indices[:max_len]
    return torch.tensor(indices, dtype=torch.long)


x = torch.stack([sentence_to_token(sentence=sentence, vocab=vocab, max_len=max_len) for sentence in data])
y = torch.tensor(labels)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
import torch
import torch.nn as nn

class TransformerClass(nn.Module):
    def __init__(self, vocab_size, embedding_dim, num_head, num_encoder_layer, hidden_dim, num_classes, max_len=512):
        super(TransformerClass, self).__init__()

        self.embedding = nn.Embedding(num_embeddings=vocab_size, embedding_dim=embedding_dim)

        self.positionalEncoding = nn.Parameter(torch.randn(1, max_len, embedding_dim))

        encoder_layer = nn.TransformerEncoderLayer(d_model=embedding_dim, nhead=num_head, dim_feedforward=hidden_dim)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=num_encoder_layer)

        self.max_len = max_len
        self.fc = nn.Linear(in_features=embedding_dim * max_len, out_features=hidden_dim)
        self.out = nn.Linear(in_features=hidden_dim, out_features=num_classes)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        if x.dim() == 1:
            x = x.unsqueeze(0)  # [seq_len] -> [1, seq_len]

        batch_size, seq_len = x.shape

        if seq_len > self.max_len:
            raise ValueError(f"Giriş sekans uzunluğu ({seq_len}) 'max_len' ({self.max_len}) değerini aşamaz.")

        embedding = self.embedding(x)
        pos_enc = self.positionalEncoding[:, :seq_len, :]
        embedded = embedding + pos_enc

        embedded = embedded.permute(1, 0, 2)
        output = self.transformer(embedded)
        output = output.permute(1, 0, 2)
        output = output.reshape(batch_size, -1)
        output = torch.relu(self.fc(output))
        output = self.out(output)
        return self.sigmoid(output)

```

```python
vocab_size = len(vocab)
embedding_dim=32
num_head=4
num_encoder_layer=4
hidden_dim=64
num_classes=1
model = TransformerClass(vocab_size=vocab_size, embedding_dim=embedding_dim, num_head=num_head, num_encoder_layer=num_encoder_layer, hidden_dim=hidden_dim, num_classes=num_classes, max_len=max_len)
```

```python
loss_fn = nn.BCELoss()
optimizer = optim.Adam(params=model.parameters(), lr=1e-3) 
```

```python
epochs = 10
model.train()

for epoch in range(epochs):
    optimizer.zero_grad()
    output = model(x_train.long()).squeeze()
    loss = loss_fn(output, y_train.float())
    loss.backward()
    optimizer.step()

    print(f"Epoch: {epoch}, Loss: {loss:.3f}")
```

```python
from sklearn.metrics import accuracy_score

# Test verisi ile tahmin
model.eval()
with torch.no_grad():
    y_pred = model(x_test.long())           # [batch_size, 1]
    y_pred = y_pred.squeeze(1)              # [batch_size]
    y_pred = (y_pred > 0.5).float()         # Binary threshold

# Sklearn ile değerlendirme
y_pred_np = y_pred.cpu().numpy()
y_test_np = y_test.float().cpu().numpy()

from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test_np, y_pred_np)
print(f"Accuracy: {acc:.4f}")

```
